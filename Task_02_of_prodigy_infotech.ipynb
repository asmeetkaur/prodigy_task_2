```python
import pandas as pd

# Load the dataset
titanic_data = pd.read_csv('/content/train.csv')

# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
# Display information about the dataset
print(titanic_data.info())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())
```

```python
# Fill missing values in the 'Age' column with the median age
median_age = titanic_data['Age'].median()
titanic_data['Age'].fillna(median_age, inplace=True)

# Fill missing values in the 'Embarked' column with the most common value
most_common_embarked = titanic_data['Embarked'].mode()[0]
titanic_data['Embarked'].fillna(most_common_embarked, inplace=True)

# Drop the 'Cabin' column due to a large number of missing values
titanic_data.drop('Cabin', axis=1, inplace=True)

# Verify that missing values have been handled
print(titanic_data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Visualize the distribution of survival (target variable)
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribution of Survival')
plt.show()
```

```python
# Explore the relationship between survival and other variables
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(x='Age', data=titanic_data, bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=titanic_data)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```
